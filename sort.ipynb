```python
def SortFile(inp, out='buf_sorted', local=True):
    print('sorting file ' + inp)
    with open(inp) as sr:
        arr = list(map(int, sr.read().split('\n')[:-1]))
    start = time.monotonic()
    arr = LibrarySort(arr, False)#True - local; False - total
    end = time.monotonic()
    print('\t', end - start)
    with open(out, 'w') as sw:
        [sw.write(str(i)+'\n') for i in arr]
    #print('stop sort file ' + inp)
    return float(end - start)
```

```python
def SplitFile(inp, maxCount):
    print('start split file ' + inp)
    fCount = -1
    curCount = 0
    numbers = []
    files = []
    with open(inp) as sr:
        for line in sr:
            if (curCount == maxCount):
                fCount += 1
                files.append(str(fCount))
                with open(str(fCount), 'w') as sw:
                    [sw.write(str(i)+'\n') for i in numbers]
                curCount = 0
                numbers = []
            numbers.append(int(line.strip()))
            curCount += 1
    fCount += 1
    files.append(str(fCount))
    with open(str(fCount), 'w') as sw:
         [sw.write(str(i)+'\n') for i in numbers]
    print('stop split file ' + inp)
    return files
```

```python
def Merge(out, files):
    print('------ merge files -------')
    readers = []
    values = []
    for f in files:
        readers.append(open(f))
        values.append(int(readers[-1].readline().strip()))
    with open(out, 'w') as sw:
        while(len(readers) > 0):
            ind = values.index(min(values))
            sw.write(str(values[ind])+'\n')
            t = readers[ind].readline().strip()
            if (t == ''):
                print('end merge file')
                del readers[ind]
                del values[ind]
            else:
                values[ind] = int(t)
```

```python
import time
from libSort import LibrarySort
def MergeLibSort(inp, out):
    print('\t\t\tstart sorting')
    files = SplitFile(inp, 10000)
    t = 0
    for f in files:
        t += SortFile(f, f)
    Merge(out, files)
    print('\t\t\tend sorting ')
    print(t)
MergeLibSort('input', 'out')
```
